# Section 1: Dataset(sample)

```python
import pandas as pd

# creating a basic sample dataset
data = {'Month':['Jan','Feb','Mar','Apr','May'],
        'Sales': [150, 200, 180, 220, 250],
        'Expenses': [80, 100, 90, 110, 120]
}

df = pd.DataFrame(data)
df
```

# Section 2: Line Plots - The classic visualization

```python
import matplotlib.pyplot as plt

df['Month']

plt.plot(df['Month'], df['Sales'], marker='o', label='Sales')
plt.plot(df['Month'], df['Expenses'], marker='o', label='Expenses')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Monthly Sales and Expenses')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5, 3))
plt.plot(df['Month'], df['Sales'], color='red', linestyle='--', marker='o', label='Sales', 
         alpha=0.5)
plt.plot(df['Month'], df['Expenses'], color='blue', marker='o', label='Expenses')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('What do you want?')
plt.legend()
plt.show()
```

# Section 3: Bar Plots - Effective for comparisons

```python
plt.figure(figsize=(4,3))
plt.bar(df['Month'], df['Sales'], color='skyblue', label='Sales')
plt.bar(df['Month'], df['Expenses'], label='Expenses')
plt.legend()
plt.show()

df
```

```python
bar_width = 0.35
bar_positions = range(len(df))

sales_x_pos = [pos for pos in bar_positions]
exp_x_pos = [pos + bar_width - 0.1 for pos in bar_positions]

print(f'{sales_x_pos = }')
print(f'{exp_x_pos = }')

plt.figure(figsize=(4,3))
plt.bar(sales_x_pos, df['Sales'], width=bar_width, label='Sales')
plt.bar(exp_x_pos, df['Expenses'], width=bar_width, label='Expenses')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.barh(df['Month'], df['Sales'])
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.scatter(df['Sales'], df['Expenses'])
plt.xlabel('Sales')
plt.ylabel('Expenses')
plt.show()
```
